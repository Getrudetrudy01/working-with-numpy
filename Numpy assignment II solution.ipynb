```python
#question 3
#Write a NumPy program to create a random array with 1000 elements and compute the average, variance, standard deviation of the array elements. 
#creating the random array using random rand
random_array = np.random.rand(1000)
```

```python
#question 4
#Write a NumPy program to calculate cumulative sum of the elements along a given axis, sum over rows for each of the 3 columns and sum over columns for each of the 2 rows of a given 3x3 array. 
#creating a 3x3 array
arrayy = np.array([[3,2,1],[6,5,4],[9,8,7]])
```

```python
arrayy
```

```python
#calaculating the cummulative sum of elements along a given axis (assume axis = 0)
cummusum_rows = np.cumsum(arrayy,axis=0)
```

```python
cummusum_rows
```

```python
# calaculating the cummulative sum of elements along a given axis(assume axis = 1)
cummumsum_col = np.cumsum(arrayy,axis=1)
cummumsum_col
```

```python
#calculating the sum over rows for each of the 3 columns
sum_over_column = np.sum(arrayy, axis=0)
sum_over_column
```

```python
#sum over columns for each of the 2 rows
sum_over_rows = np.sum(arrayy, axis = 1)
sum_over_rows
```
